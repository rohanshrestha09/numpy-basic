```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
plt.plot(yield_apples, marker='o')
```

```python
years = range(2010, 2016)
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
plt.plot(years, yield_apples, marker='o')
```

```python
plt.plot(years, yield_apples)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
```

```python
years = range(2000, 2010)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.937]
orange = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898]
plt.plot(years, apples)
plt.plot(years, orange)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
```

```python
plt.plot(years, apples, marker="o")
plt.plot(years, orange, marker="x")
plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")
plt.legend(["Apples", "Oranges"])
plt.title("Crop Yields in Kanto")
```

```python
sns.set_style("whitegrid")
plt.plot(years, apples, 's-b')
plt.plot(years, orange, 'o--r')

plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")

plt.title("Crop Yields in Kanto")
plt.legend(["Apples", "Oranges"])
```

```python
sns.set_style("darkgrid")

plt.plot(years, apples, 's-b')
plt.plot(years, orange, 'o--r')

plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")

plt.title("Crop Yields in Kanto")
plt.legend(["Apples", "Oranges"])
```
